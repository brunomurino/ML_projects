# Binary Logistic Regression

We perform a Logistic Regression when we have a set of features $x$ we want to map to some label $y$, but we can't (or don't want to) explicitly find a solution.

A Binary Logistic Regression takes as input a training set $(X,Y)$, where $ X \in \mathbb{R}^{n_x, m}$, i.e. a $m$ column vectors, called **examples**, with $n_x$ entries, called **features**.

The Cost Function, denoted by $J(\omega, b)$, is defined as
$$J(\omega, b) = \frac{1}{m}\sum_{i=1}^m L(A^{(i)}, Y^{(i)}),$$ where
$L(A, Y)$ is the Loss Function (error of single training
example), defined as $$L(A, Y) = -\Big[Y*\ln(A) + (1-Y)*\ln(1-A)\Big]$$
