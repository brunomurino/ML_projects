# Built-in functions vs For loops

GPU's and CPU's have SIMD: single instruction multiple data, which makes use of parallelism to improve performance. Built-in functions, such as Numpy's, make use of this, achieving better performance. GPU's are better at this than CPU's.

```python
import numpy as np
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()

print(c)
print("Vectorized version: " + str(1000*(toc-tic)) + "ms")

c = 0
tic = time.time()
for i in range(1000000):
    c += a[i]*b[i]
toc = time.time()

print(c)
print("Non-vectorized version: " + str(1000*(toc-tic)) + "ms")
```

# Numpy

```python
# GRADED FUNCTION: sigmoid

import numpy as np # this means you can access numpy functions by writing np.function() instead of numpy.function()

def sigmoid(x):
    """
    Compute the sigmoid of x

    Arguments:
    x -- A scalar or numpy array of any size

    Return:
    s -- sigmoid(x)
    """
    
    ### START CODE HERE ### ( 1 line of code)
    s = 1/(1+np.exp(-x))
    ### END CODE HERE ###
    
    return s

x = np.array([1, 2, 3])
print(sigmoid(x))

from bokeh.plotting import figure
from bokeh.io import show, output_notebook, reset_output
reset_output()
output_notebook()
range = 7
p = figure(x_axis_label='x', y_axis_label='sigmoid(x)')
p.line(np.linspace(-range, range, num=10000), sigmoid(np.linspace(-range, range, num=10000)))
show(p)
```

```python
# GRADED FUNCTION: sigmoid_derivative

def sigmoid_derivative(x):
    """
    Compute the gradient (also called the slope or derivative) of the sigmoid function with respect to its input x.
    You can store the output of the sigmoid function into variables and then use it to calculate the gradient.
    
    Arguments:
    x -- A scalar or numpy array

    Return:
    ds -- Your computed gradient.
    """
    
    ### START CODE HERE ### ( 2 lines of code)
    s = sigmoid(x)
    ds = s*(1-s)
    ### END CODE HERE ###
    
    return ds

x = np.array([1, 2, 3])
print ("sigmoid_derivative(x) = " + str(sigmoid_derivative(x)))

from bokeh.plotting import figure
from bokeh.io import show, output_notebook, reset_output
reset_output()
output_notebook()
range = 7
p = figure(x_axis_label='x', y_axis_label='sigmoid_derivative(x)')
p.line(np.linspace(-range, range, num=10000), sigmoid_derivative(np.linspace(-range, range, num=10000)))
show(p)
```

```python
# GRADED FUNCTION: image2vector
def image2vector(image):
    """
    Argument:
    image -- a numpy array of shape (length, height, depth)
    
    Returns:
    v -- a vector of shape (length*height*depth, 1)
    """
    
    ### START CODE HERE ### ( 1 line of code)
    v = image.reshape(image.shape[0]*image.shape[1]*image.shape[2],1)
    ### END CODE HERE ###
    
    return v
```

```python
def normalizeRows(x): # This should be used on X BEFORE transposing it in order to normalize each example
    """
    Implement a function that normalizes each row of the matrix x (to have unit length).
    
    Argument:
    x -- A numpy matrix of shape (n, m)
    
    Returns:
    x -- The normalized (by row) numpy matrix. You are allowed to modify x.
    """

    # Compute x_norm as the norm 2 of x. Use np.linalg.norm(..., ord = 2, axis = ..., keepdims = True)
    x_norm_row = np.linalg.norm(x, ord = 2, axis = 1, keepdims = True)
    
    # To compute the norm by column
    x_norm_column = np.linalg.norm(x, ord = 2, axis = 0, keepdims = True)
    
    # Divide x by its norm.
    x = x/x_norm_row
#     x = x/x_norm_column

    return x
```
