# UK Map visualisations

The goal of this notebook is to sucessfully make a geographic plot of the boroughs of London.

Let's start by importing pandas:

````python
import pandas as pd
````

You can download [here](https://www.doogal.co.uk/UKPostcodes.php) a csv table with the coordinates for each postcode in the UK, and read some of it's columns with: 

````python
postcode_file = 'postcodes.csv'

postcodes = pd.read_csv(postcode_file, usecols=[
    'Postcode', 
    'Latitude', 
    'Longitude',
    'County', 
    'District',
    'Region'], low_memory=False)
````

Let's see how many postcodes belong to the district of City of London:

````python
len(postcodes[postcodes['District'] == 'City of London'])
````

Our goal is to find out how many postcodes each District from Greater London has. In order to find that we're going to use only the postcodes from the region of London and group the data by District, counting the number of postcodes and computing the mean latitude and longitude of each borough.

````python
def f(df):
    d = {}
    d['#ofPostcodes'] = df['Latitude'].count()
    d['Latitude'] = df['Latitude'].mean()
    d['Longitude'] = df['Longitude'].mean()
    return pd.Series(d, index=['#ofPostcodes', 'Latitude', 'Longitude'])

counties = postcodes[postcodes['Region']=='London'].groupby('District').apply(f)
````

Let's see what we've got so far

````python
counties.sort_values(by='#ofPostcodes', ascending=False)
````

Here we can see that the district of Westminster is the one with more postcodes, with nearly twice as many as the second place. Let's do a geographic plot to see these results in a map on Greater London.

## Getting the files for the geographic plot

Download the files from [here](https://data.london.gov.uk/download/statistical-gis-boundary-files-london/9ba8c833-6370-4b11-abdc-314aa020d5e0/statistical-gis-boundaries-london.zip). The files we'll be using are in the folder `ESRI` and are named `London_Borough_Excluding_MHW`, with various extensions. I moved this files to a folder called `shapefiles`, in the same directory as this notebook.

### Converting the files

In order to use the files we downloaded we need to convert them to a slightly different format. To do so, in the terminal go to the shapefiles directory and run:
```
$ ogr2ogr -t_srs EPSG:4326 -f "ESRI Shapefile" London_Borough_Excluding_MHW_WGS84.shp London_Borough_Excluding_MHW.shp
```
The above command will take all the files named `London_Borough_Excluding_MHW` and convert them to new files named `London_Borough_Excluding_MHW_WGS84`. Note that the command automatically converts the files with other extensions, not just the one `.shp`.

## Plotting the map
First, we import the libraries:

````python
import matplotlib.pyplot as plt
import matplotlib.cm

from mpl_toolkits.basemap import Basemap
# for high or full resolution you need to run: ! conda install -c conda-forge basemap-data-hires --yes
````

Then we define the actual region we want to display, using coodinates:

````python
lat_centre = 51.5 # This is the latitude of the center of our region
lon_centre = -0.128 # And this is the corresponding longitude

offset_west = 0.4 # This is how many degrees to the west we want to display, and analogous for the others.
offset_east = 0.47
offset_south = 0.23
offset_north = 0.2
````

And finally we do the plot:

````python
fig, ax = plt.subplots(figsize=(50,50))

map = Basemap(resolution='c', # c, l, i, h, f or None
            projection='merc',
            lat_0 = lat_centre,
            lon_0 = lon_centre, # latitude and longitude of the center of the map
            llcrnrlon = lon_centre - offset_west,
            llcrnrlat = lat_centre - offset_south, # latitude and longitude of the lower left corner of the map
            urcrnrlon = lon_centre + offset_east,
            urcrnrlat = lat_centre + offset_north) # latitude and longitude of the upper right of the map

map.readshapefile('shapefiles/London_Borough_Excluding_MHW_WGS84', 'london_boroughs')

def plot_area_mean(row):
    """Plot the data for a row"""
    size = row[0]/700 # Here we simply scale the count of postcodes to a reasonable number to use as markersize
    x, y = map(row[2], row[1])
    map.plot(x, y, 'o', markersize=size+5, color='#CF142B', alpha=1)
    plt.text(x - 3000, y + 1000, row.name, fontsize=25, color='#00247D', fontweight='bold')
            
_ = counties.apply(plot_area_mean, axis=1)

plt.show()
````
